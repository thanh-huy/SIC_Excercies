Tiền xử lý dữ liệu với bộ dataset iris

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
#iris['data']
iris_df = pd.DataFrame(iris['data'], columns = iris['feature_names'])
iris_df.head()
#iris_df['sepal length (cm)']
```

```python
import numpy as np
np.mean(iris_df['sepal length (cm)'])
iris_df_target = pd.DataFrame(iris['target'], columns = ['species'])
iris_df_target.head()
iris_df_target['species'].replace({0:'sentosa',1:'versicolor',2:'virginica'}, inplace = True)
iris_df_target
```

Chuyển giá trị riêng của các biến phân loại thành biến giả

```python
pd.get_dummies(iris_df_target['species'])
```

Mã hóa One-hot

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder()
len_iris = len(iris_df_target['species'])
np.array(iris_df_target['species'])
#one_hot_encoder.fit_transform(np.array(iris_df_target['species']).reshape(len_iris,1)).toarray()
iris_species_2d = np.array(iris_df_target['species']).reshape(len_iris,1)
one_hot_reshape = one_hot_encoder.fit_transform(iris_species_2d)
print(one_hot_reshape[0:50])
```
