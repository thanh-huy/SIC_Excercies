# 1. Phân cụm với K-means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
%matplotlib inline
```

## 1.1. Sinh dữ liệu từ make_blobs

```python
X1, label1 = make_blobs(n_samples=100,n_features=2,centers=2,cluster_std=5,random_state=123)
plt.scatter(X1[:,0],X1[:,1], c=label1, alpha=0.7)
plt.title('Dữ liệu sinh ngẫu nhiên 2 cụm')
plt.show()
```

```python
#Dataset 2
X2, label2 = make_blobs(n_samples=100,n_features=2,centers=3,cluster_std=1,random_state=321)
plt.scatter(X2[:,0],X2[:,1], c=label2, alpha=0.7)
plt.title('Dữ liệu sinh ngẫu nhiên 3 cụm')
plt.show()
```

## 1.2. Phân cụm với k-means

```python
kmeans = KMeans(n_clusters=2, random_state=123, n_init='auto')
kmeans.fit(X1)
myColors = {0:'red',1:'green',2:'blue',3:'yellow'}
plt.scatter(X1[:,0],X1[:,1], c = pd.Series(kmeans.labels_).apply(lambda x:myColors[x]), alpha=0.7)
plt.title('Kết quả phân cụm')
plt.show()
```

```python
#Phân cụm Dataset 1 với 3 clusters
kmeans = KMeans(n_clusters=4, random_state=123, n_init='auto')
kmeans.fit(X1)
plt.scatter(X1[:,0],X1[:,1], c = pd.Series(kmeans.labels_).apply(lambda x:myColors[x]), alpha=0.7)
plt.title('Kết quả phân cụm')
plt.show()
```

```python
#Phân cụm với dataset2
kmeans = KMeans(n_clusters=4, random_state=321, n_init='auto')
kmeans.fit(X2)
plt.scatter(X2[:,0],X2[:,1], c = pd.Series(kmeans.labels_).apply(lambda x:myColors[x]), alpha=0.7)
plt.title('Kết quả phân cụm')
plt.show()
```
