# Các phép toán trên Ma trận

```python
import numpy as np
```

```python
arr = np.random.randn(6,3) # Khởi tạo ma trận 6x3 ngẫu nhiên
arr
```

```python
arr.T #Chuyển vị của ma trận A
```

```python
#Nhân ma trận không sử dụng toán tử *
#Sử dụng np.dot(A,B)
np.dot(arr.T,arr)
```

```python
arr1 = np.arange(12).reshape(3,4)
arr1
```

```python
arr1.reshape(4,3)
```

```python
arr1.transpose(0,1)
```

```python
arr1.transpose(1,0)
```

```python
arr = np.arange(16).reshape((2,2,4))
arr
```

```python
arr.transpose((1,0,2))
```

```python
arr.transpose(0,1,2)
```

```python
arr.transpose(2,1,0)
```

# Dùng hàm hoán đổi trục bằng swapaxes

```python
arr.swapaxes(1,2)
```

```python
arr.swapaxes(0,1)
```
