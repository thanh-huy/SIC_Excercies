# Mảng String và Boolean

```python
names = np.array(['Bob','Joe','Tom','Bob','Tom','Joe','Joe'])
names
```

```python
names == 'Bob' #Kiểm tra các phần từ có giá trị bằng  'Bob'
```

```python
names[names=='Bob'] # Lọc các phần tử có giá trị bằng  'Bob'
```

```python
arr = np.empty((8,4)) # Tạo mảng 8x4
```

```python
for i in range(8): # Gán giá trị cho mảng bằng các chỉ số hàng.
  arr[i] = i
arr
```

```python
arr[[4,3,0,6]] #Tách các hàng 4,3,0,6, đi từ trên xuống
```

```python
arr[[-3,-5,-7]] #Tách các hàng -3,-5,-7 đi từ cuối lên.
```

```python
arr2 = np.arange(20)
```

```python
arrMask = (arr2%5==0)
arrMask
```

```python
#Dùng Mask lọc trên arr2
arr2[arrMask] #Lọc ra các phần tử ở vị trí True trong Mask
```

```python
arr2>10
```

```python
arrMask2 = (arr2>=10)
arr2[arrMask & arrMask2] # Kết hợp điều kiện lọc theo Mask và Mask2
```
