```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
type(load_breast_cancer())
cancer_data = load_breast_cancer()
pd.DataFrame(cancer_data['data'])
pd.DataFrame(cancer_data['target'])
print(cancer_data.DESCR)
```

Xem tổng số lượng bản ghi để chi tỉ lệ Train_Test

```python
len(cancer_data.data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, cancer_data.target, test_size=0.3)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Huấn luyện mô hình với bộ phân lớp DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X_train, y_train)
```

Dự đoán mô hình

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
print(y_pred)
```

Chuyển output dự đoán về dạng mảng 1 chiều

```python
pred_2d = y_pred.reshape(len(y_pred),1)
print(pred_2d)
```

Tạo DataFrame so sánh kết quả dự đoán và Y_test

```python
import numpy as np
y_test_2d = y_test.reshape(len(y_test),1)
df1 = pd.DataFrame(pred_2d)
df2 = pd.DataFrame(y_test_2d)
pd.concat([df1,df2], axis=1)
```

Thêm cột tiêu đề

```python
df = pd.concat([df1, df2], axis=1)
df.columns = ['y_pred', 'y_test']
df.columns
#So sanh từng hàng trong các cột
df[df['y_pred']==df['y_test']]
len(df[df['y_pred']==df['y_test']])/len(df)
```
